```python
import pandas as pd
from pathlib import Path
import ast
```

```python
BACDIVE_DIR = Path.cwd().parent / "kg_microbe/transform_utils/bacdive"
```

```python
df = pd.read_csv(BACDIVE_DIR/"tmp/bacdive.tsv", sep="\t", low_memory=False)
df.head()
```

```python
# Step 1: Convert stringified lists to actual lists
df['keywords'] = df['keywords'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else x)

# Ensure all elements are lists (for any remaining single string entries)
df['keywords'] = df['keywords'].apply(lambda x: x if isinstance(x, list) else [x])

# Step 2: Explode the 'keywords' column
exploded_df = df.explode('keywords')

# Step 3: Get unique values
unique_keywords = exploded_df['keywords'].unique()

# Step 4: Convert to list
unique_keywords_list = unique_keywords.tolist()

# Display the list of unique keywords
print(unique_keywords_list)
```

```python
df2 = pd.read_csv(BACDIVE_DIR/"tmp/bacdive_physiology_metabolism.tsv", sep="\t", low_memory=False)
df2.head()
```

```python
o2_tolerance_filtered_df = df2[df2['oxygen tolerance'].notna()]
o2_tolerance_filtered_df.head()
```

```python
# Function to extract 'oxygen tolerance' from a dict or list of dicts
def extract_oxygen_tolerance(data):
    if isinstance(data, dict):
        # If data is a single dictionary, return the 'oxygen tolerance' value
        return data['oxygen tolerance']
    elif isinstance(data, list):
        # If data is a list of dictionaries, extract 'oxygen tolerance' from each dict
        return [d['oxygen tolerance'] for d in data]
    else:
        # Handle other unexpected data types
        return None
```

```python
o2_tolerance_series = o2_tolerance_filtered_df['oxygen tolerance']

oxygen_tolerance_values = o2_tolerance_series.apply(extract_oxygen_tolerance).explode()

# Get unique values using unique() method
unique_oxygen_tolerance_values = oxygen_tolerance_values.dropna().unique()

print(unique_oxygen_tolerance_values)
```
