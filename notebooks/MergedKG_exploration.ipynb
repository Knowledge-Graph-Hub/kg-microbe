```python
from pathlib import Path
import tarfile
import pandas as pd
```

```python
MERGED_TAR = Path.cwd().parent / "data/merged/merged-kg.tar.gz"
```

```python
tar = tarfile.open(MERGED_TAR, "r:gz")
# Extract the TSV file
for member in tar.getmembers():
    f = tar.extractfile(member)
    if '_nodes.tsv' in member.name:
        # Read the TSV file into a pandas DataFrame
        df_nodes = pd.read_csv(f, sep='\t', low_memory=False)
    elif '_edges.tsv' in member.name:
        # Read the TSV file into a pandas DataFrame
        df_edges = pd.read_csv(f, sep='\t', low_memory=False)

# Close the tar.gz file
tar.close()
```

```python
subj_pattern = ['medium:', 'solution:', 'CHEBI']
obj_pattern = ['solution:', 'ingredient:', 'CHEBI']

condition_1 = df_edges['subject'].str.contains('|'.join(subj_pattern))
condition_2 = df_edges['object'].str.contains('|'.join(obj_pattern))

df_mediadive = df_edges[condition_1 & condition_2]

# mapping = df_edges.set_index('object')['subject'].to_dict()

# mapping
df_mediadive
```

```python
medium_subject_condition = df_mediadive["subject"].str.startswith("medium:")
solution_object_condition = df_mediadive["object"].str.startswith("solution:")

medium_solution_df = df_mediadive[medium_subject_condition & solution_object_condition]
medium_solution_df
```

```python
solution_subject_condition = df_mediadive["subject"].str.startswith("solution:")
ingredient_object_condition = df_mediadive["object"].str.startswith("ingredient:")

solution_ingredient_df = df_mediadive[solution_subject_condition & ingredient_object_condition]
solution_ingredient_df
```

```python
chebi_subject_condition = df_mediadive["subject"].str.startswith("CHEBI:")
chebi_object_condition = df_mediadive["object"].str.startswith("CHEBI:")

chebi_chebi_df = df_mediadive[chebi_subject_condition & chebi_object_condition]
# medium_solution_df

chebi_chebi_df
```

```python
chebi_subject_df  = df_mediadive[chebi_subject_condition & ~chebi_object_condition]
chebi_subject_df
```

```python
chebi_object_df  = df_mediadive[~chebi_subject_condition & chebi_object_condition]
chebi_object_df
```
