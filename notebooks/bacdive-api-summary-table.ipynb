## Testing Bacdive API

```python
import bacdive
import pandas as pd
import numpy as np
```

```python
with open('up.txt', 'r') as file:
    creds = [line.rstrip() for line in file]
```

```python
client = bacdive.BacdiveClient(creds[0], creds[1])
```

```python
##read bacdive strains data file -- downloaded from website after filtering for species
df = pd.read_csv('strains_advsearch_bacdive_2023-03-10.csv')
df.head()
```

```python
df.info()
```

```python
len(df.species.unique())
```

```python
##get unique species names
species_all = df.species.unique().tolist()
len(species_all)
```

```python
import time
```

```python
##Search the filtered data for each unique species name in Bacdive and return NCBI ID, physiology & metabolism, and isolation data
filter = ['NCBI tax id', 'Physiology and metabolism', 'Isolation, sampling and environmental information']
bacdive_data = {}
count = 0
error = []
for species_name in species_all:
    bacdive_data[species_name] = []
    try:
        query = {'taxonomy': species_name}
        client.search(**query)
        result = client.retrieve(filter)
        for item in result:
            bacdive_data[species_name].append(item)
        #time.sleep(3)
    except Exception as e:
        error.append(species_name)
    count += 1
    if count%1000 == 0:
        print(count)
len(bacdive_data)
```

```python
len(error)
```

```python
##save data dictionary
import pickle
with open('bacdivedata_ID_isolation_metabolism_20230321.pickle', 'wb') as filep:
    pickle.dump(bacdive_data, filep)
```
